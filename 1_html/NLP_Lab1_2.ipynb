```python
import random
import nltk
from nltk.corpus import gutenberg, brown
from nltk import ConditionalFreqDist

nltk.download('brown')
```

```python
text = brown.words(categories='news')
print(text)
```

```python
bigrams = nltk.bigrams(text)
cfd = ConditionalFreqDist(bigrams)

def generate_random_text(cfdist, start_word, num_words=50, n=3):
    word = start_word
    output = [word]

    for _ in range(num_words - 1):
        if word not in cfdist:
            break

        most_likely_words = [w for w, _ in cfdist[word].most_common(n)]

        word = random.choice(most_likely_words)
        output.append(word)

    return ' '.join(output)
```

```python
for n_val in [3, 5, 10]:
    print(generate_random_text(cfd, start_word='The', num_words=50, n=n_val))
```
